```python
from sklearn.datasets import make_classification


```

```python
X,y=make_classification(n_samples=1000,n_features=2,n_redundant=0,n_clusters_per_class=1,weights=[0.90],random_state=12
)
```

```python
X.shape
```

```python
y.shape
```

```python

import numpy as np
import pandas as pd

df1=pd.DataFrame(X,columns=['f1','f2'])
df2 = pd.DataFrame(y,columns=['target'])
df=pd.concat([df1,df2],axis=1)
df.head(20)
```

```python
df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['f1'],df['f2'],c=df['target'],cmap='coolwarm')
plt.xlabel('f1')
plt.ylabel('f2')
plt.title('Scatter plot of f1 and f2')
plt.show()
```

```python
from imblearn.over_sampling import SMOTE
```

```python
oversample = SMOTE()
X, y = oversample.fit_resample(df[['f1', 'f2']], df['target'])
df = pd.DataFrame(X, columns=['f1', 'f2'])
```

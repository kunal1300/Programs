```python
df['sales_nan']= df['Sales'].fillna(df['Sales'].mean())
df['value_nan']= df['Value'].fillna(df['Value'].mean())
df
```

```python
# renaming column name
df=df.rename(columns={'Date': 'Sales_date'})
df
```

```python
# convert into float to integer value

df['value_nan']= df['Value'].fillna(df['Value'].mean()).astype(int)
df['sales_nan']=df['Sales'].fillna(df['Sales'].mean()).astype(int)
df['Sales']= df['Sales'].fillna(df['Sales'].mean()).astype(int)
df['Value']= df['Value'].fillna(df['Value'].mean()).astype(int)
df
```

```python
df['New_value'] = df['Value'].apply(lambda x:x*2)
df 
```

```python
#group by
group_mean = df.groupby('Product')['Value'].mean()
group_mean

```

```python
# merging and joining dataframe

df11 =pd.DataFrame({'key':['A','B','C','D'],'Value1':[1,2,3,4]})
df12 =pd.DataFrame({'key':['A','B','C','F'],'Value1':[4,2,6,4]})
merge_df= pd.merge(df11,df12,on='key',how='inner')
merge_df
```

```python
merge_df= pd.merge(df11,df12,on='key',how='outer')
merge_df
```

```python
merge_df= pd.merge(df11,df12,on='key',how='left')
merge_df
```

```python
merge_df= pd.merge(df11,df12,on='key',how='right')
merge_df
```
